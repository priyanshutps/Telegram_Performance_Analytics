```python
import csv
import matplotlib.pyplot as plt

# Load data from CSV file
def load_data(filename):
    weeks, clicks, earnings, posts = [], [], [], []
    with open(filename, 'r') as f:
        reader = csv.reader(f)
        next(reader)  # skip header
        for row in reader:
            weeks.append(row[0])
            clicks.append(int(row[1]))
            earnings.append(int(row[2]))
            posts.append(int(row[3]))
    return weeks, clicks, earnings, posts

# Calculate earnings per post and per 1000 clicks
def efficiency_analysis(weeks, clicks, earnings, posts):
    earnings_per_post = []
    earnings_per_1k_clicks = []

    for i in range(len(weeks)):
        # Earnings per post
        if posts[i] > 0:
            epp = earnings[i] / posts[i]
        else:
            epp = 0
        earnings_per_post.append(epp)

        # Earnings per 1000 clicks
        if clicks[i] > 0:
            epc = earnings[i] / (clicks[i] / 1000)
        else:
            epc = 0
        earnings_per_1k_clicks.append(epc)

    return earnings_per_post, earnings_per_1k_clicks

def plot_earnings(weeks, earnings):
    plt.figure(figsize=(12, 6))
    plt.bar(weeks, earnings, color='skyblue')
    plt.xticks(rotation=45)
    plt.title("Weekly Earnings (Telegram Channel)")
    plt.xlabel("Week")
    plt.ylabel("Earnings (‚Çπ)")
    plt.tight_layout()
    plt.show()

# Find top and bottom weeks
def find_extremes(weeks, earnings, posts, clicks):
    max_index = earnings.index(max(earnings))
    min_index = earnings.index(min(earnings))
    return {
        'best': (weeks[max_index], earnings[max_index], posts[max_index], clicks[max_index]),
        'worst': (weeks[min_index], earnings[min_index], posts[min_index], clicks[min_index])
    }

# Detect underperforming weeks
def detect_weak_weeks(weeks, earnings, posts, clicks):
    weak_weeks = []
    for i in range(len(weeks)):
        if earnings[i] < 1000 and posts[i] >= 3:
            weak_weeks.append((weeks[i], earnings[i], posts[i], clicks[i]))
    return weak_weeks

# Recommend insights based on data patterns
def generate_tips(weeks, posts, earnings):
    tips = []
    high_post_weeks = [i for i in range(len(posts)) if posts[i] >= 6]
    if max(posts) > 6:
        tips.append("Earning per post increases with higher post volume.")
    low_post_under_4 = [i for i in range(len(posts)) if posts[i] < 20 and earnings[i] < 1000]
    if len(low_post_under_4) >= 1:
        tips.append("Avoid weeks with fewer than 20 posts ‚Äî they often underperform.")
    return tips

# Main program
weeks, clicks, earnings, posts = load_data("telegram_data.csv")
epp, epc = efficiency_analysis(weeks, clicks, earnings, posts)
extremes = find_extremes(weeks, earnings, posts, clicks)
weak_weeks = detect_weak_weeks(weeks, earnings, posts, clicks)
tips = generate_tips(weeks, posts, earnings)

# Display Results
print("\nüìä Telegram Weekly Performance Insights\n")
print(f"‚úÖ Best Week: {extremes['best'][0]} ‚Äì ‚Çπ{extremes['best'][1]} (Posts: {extremes['best'][2]}, Clicks: {extremes['best'][3]})")
print(f"‚ö†Ô∏è  Worst Week: {extremes['worst'][0]} ‚Äì ‚Çπ{extremes['worst'][1]} (Posts: {extremes['worst'][2]}, Clicks: {extremes['worst'][3]})")

# Show most efficient week
most_efficient_index = epp.index(max(epp))
most_click_efficient_index = epc.index(max(epc))
print(f"\nüí° Highest Earnings per Post: {weeks[most_efficient_index]} ‚Äì ‚Çπ{round(epp[most_efficient_index], 2)}/post")
print("   ‚Ü™ You earned the most per post this week.")
print("   üí¨ Advice: Analyze what you posted that week.")
print("      ‚Ä¢ Was it high-converting content?")
print("      ‚Ä¢ Did you post during peak hours?")
print("      ‚Ä¢ Were the offers urgent or limited-time?")
print(f" \nüí° Best Click Efficiency: {weeks[most_click_efficient_index]} ‚Äì ‚Çπ{round(epc[most_click_efficient_index], 2)} per 1000 clicks")
print("   ‚Ü™ Your audience engaged and converted better than usual.")
print("   üí¨ Advice: Review what content earned you more per click.")
print("      ‚Ä¢ Was the call-to-action strong?")
print("      ‚Ä¢ Was the product more profitable?")
print("      ‚Ä¢ Did you use better link placement or formatting?")

# Show underperforming weeks
if weak_weeks:
    print("\n‚ö†Ô∏è Underperforming Weeks:")
    for week, earn, post, click in weak_weeks:
        print(f" - {week}: ‚Çπ{earn}, Posts: {post}, Clicks: {click}")
    print("\nüí¨ Advice for Underperforming Weeks: Review what you posted during these weeks.")
    print("   ‚Ü™ Was the content less engaging?")
    print("   ‚Ü™ Were titles or photos weak?")
    print("   ‚Ü™ Did you post less frequently?")
    print("   ‚Ü™ Were there fewer trending offers?")
    print("üëâ Use this info to learn what not to repeat.")
    
# Show tips
if tips:
    print("\nüìå Tips from Data:")
    for tip in tips:
        print(f" - {tip}")

plot_earnings(weeks, earnings)
```

```python
from bs4 import BeautifulSoup
from datetime import datetime
import os # To check if files exist

def extract_data_simply_from_html(html_content):
    """
    Extracts date, time, and categorizes time from HTML content simply.
    Outputs a list of lists.
    """
    soup = BeautifulSoup(html_content, 'html.parser') # Using default parser
    
    data_rows = []
    current_date = None

    # Iterate through all 'div' elements to find message blocks
    for block in soup.find_all('div'):
        block_classes = block.get('class', [])

        # Check for date message (e.g., "30 April 2025")
        if 'message' in block_classes and 'service' in block_classes:
            date_text_element = block.find('div', class_='body details')
            if date_text_element:
                date_text = date_text_element.text.strip()
                try:
                    # Parse date format "DD Month YYYY"
                    current_date = datetime.strptime(date_text, '%d %B %Y').strftime('%d-%m-%Y') 
                except ValueError:
                    # Handle cases where month might be short (e.g., 'Apr' instead of 'April')
                    try:
                        current_date = datetime.strptime(date_text, '%d %b %Y').strftime('%d-%m-%Y') 
                    except ValueError:
                        current_date = None # Reset if date format is not recognized

        # Check for default message containing a timestamp
        elif 'message' in block_classes and 'default' in block_classes:
            time_div = block.find('div', class_='pull_right date details')
            if time_div:
                title = time_div.get('title')
                if title and current_date: # Ensure we have a date to associate with
                    try:
                        # The title format is "DD.MM.YYYY HH:MM:SS UTC+05:30"
                        full_datetime_str = title.split(' UTC')[0]
                        dt_object = datetime.strptime(full_datetime_str, '%d.%m.%Y %H:%M:%S')

                        extracted_time = dt_object.strftime('%H:%M')
                        hour = dt_object.hour
                        
                        # --- MODIFIED LOGIC FOR TIME_CATEGORY (Morning/Afternoon/After Evening) ---
                        if (hour >= 17) or (hour >= 0 and hour < 4):
                            time_category = 'After Evening'
                        elif (hour >= 4 and hour < 12):
                            time_category = 'Morning'
                        else: # This covers 12 PM (noon) to before 5 PM (17:00)
                            time_category = 'Afternoon'
                        # --- END MODIFIED LOGIC ---

                        # Use the date from the title, as it's directly associated with the timestamp
                        extracted_date_from_title = dt_object.strftime('%d-%m-%Y')

                        data_rows.append([extracted_date_from_title, extracted_time, time_category])
                    except ValueError:
                        pass # Skip if time parsing fails
    return data_rows

# --- Main execution for Jupyter ---

file_names = ['messages.html', 'messages2.html']
all_extracted_data = []

for file_name in file_names:
    if not os.path.exists(file_name):
        print(f"Error: {file_name} not found in the current directory. Please ensure it's there.")
        continue

    try:
        with open(file_name, 'r', encoding='utf-8') as f:
            html_content = f.read()
        print(f"Successfully read {file_name}. Content length: {len(html_content)} bytes")
        
        extracted_from_file = extract_data_simply_from_html(html_content)
        all_extracted_data.extend(extracted_from_file)

    except Exception as e:
        print(f"An error occurred while processing {file_name}: {e}")

# Sort the collected data by date and time
# First, convert strings to datetime objects for accurate sorting
sortable_data = []
for row in all_extracted_data:
    try:
        # Create a datetime object for sorting - IMPORTANT: use the format that the row[0] string is in
        dt_obj = datetime.strptime(f"{row[0]} {row[1]}", '%d-%m-%Y %H:%M') 
        sortable_data.append((dt_obj, row))
    except ValueError:
        # If parsing fails for sorting, just append the row as is
        sortable_data.append((datetime.min, row)) 

sortable_data.sort(key=lambda x: x[0]) # Sort by the datetime object

# Extract only the original rows after sorting
sorted_final_data = [row for dt_obj, row in sortable_data]


# Prepare output string for CSV (comma-separated values)
# Header row
output_string = "Date,Time,Time_Category\n" 

# Data rows
for row in sorted_final_data:
    output_string += f"{row[0]},{row[1]},{row[2]}\n" 

# Specify the output file name
output_file_name = 'daily_post_data.csv' 

if sorted_final_data:
    try:
        with open(output_file_name, 'w', encoding='utf-8') as outfile:
            outfile.write(output_string)
        print(f"\nData successfully saved to '{output_file_name}'.")
        print(f"You can now open '{output_file_name}' directly in Excel, and it should separate columns automatically.")
    except Exception as e:
        print(f"\nError saving data to file '{output_file_name}': {e}")
else:
    print("\nNo data extracted from the HTML files. An output file will not be created.")
```
